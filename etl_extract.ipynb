```python
import pandas as pd

# Load datasets
raw_data = pd.read_csv('raw_data.csv')
incremental_data = pd.read_csv('incremental_data.csv')

# Preview first 5 rows
print("Raw Data Head:")
display(raw_data.head())
print("\nIncremental Data Head:")
display(incremental_data.head())

# Display basic info
print("\nRaw Data info:")
display(raw_data.info())
print("\nIncremental Data info :")
display(incremental_data.info())
```

Observations

```python
#Checking for missing values
print("Raw Data - Missing Values:")
display(raw_data.isnull().sum())

print("\nIncremental Data - Missing Values:")
display(incremental_data.isnull().sum())

```

```python
#checking for duplicates
print("Raw Data Duplicates:", raw_data.duplicated().sum())
print("Incremental Data Duplicates:", incremental_data.duplicated().sum())
```

```python
# Check for unexpected values (e.g., 'j' in numeric fields)
print("Raw Data Suspicious Values:")
display(raw_data.describe(include='all'))  # Stats for numeric + object columns

print("\nIncremental Data Suspicious Values:")
display(incremental_data.describe(include='all'))
```
